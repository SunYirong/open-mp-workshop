```python
n = 729
b = np.array([[3.142 * (i + j) for j in range(n)] for i in range(n)])

plt.figure(figsize=(10, 4))
plt.subplot(121)
sns.heatmap(b)

b2 = np.array([[(i * j + 1)/ (n **2) for j in range(n)] for i in range(n)])
plt.subplot(122)
sns.heatmap(b2)
print(a.shape, b.shape, b2.shape)
```

```python
rN2 = 1 / (n ** 2)
a = np.array([[0 for j in range(n)] for i in range(n)])
tasks1 = []
for i in range(n):
    j = n - 1
    tmp = 0
    while j > i:
        tmp += 1
        a[i, j] = a[i, j] + np.cos(b[i,j])
        j -= 1
    tasks1.append(tmp)
tasks1 = np.array(tasks1)

c = [0 for i in range(n)]
jmax = np.array([1 if (i % (3 * (i / 30) + 1)) != 0 
                     else n 
                         for i in range(n)])
tasks2 = []
for i in range(n):
    tmp = 0
    for j in range(jmax[i]):
        tmp += j
        for k in range(j):
            c[i] += (k + 1) * np.log(b2[i, j]) * rN2
    tasks2.append(tmp)
c = np.array(c)
tasks2 = np.array(tasks2)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.plot(range(n), tasks1)
plt.title("Loop 1 work load")
plt.ylabel("actions count per 'i'")
plt.xlabel("loop variable 'i'")
plt.subplot(122)
plt.plot(range(n), tasks2)
plt.title("Loop 2 work load")
plt.ylabel("actions count per 'i'")
plt.xlabel("loop variable 'i'")
```

```python
six = data[(data.threads_num==6)&(data.chunk_size!=0)]
plt.figure(figsize=(12, 4))
ax = sns.barplot(x="chunk_size", 
                 y="total_time", 
                 hue="sched_type", 
                 data=six, 
                 alpha=0.7)
label = six[(six.chunk_size==8)&(six.type==1)].iloc[0].total_time
ax.text(2.89,
        y=label+0.3,
        s=round(label, 1),
        fontsize=12)
```

```python
dynamic8 = data[(data.chunk_size==8)&(data.type==1)]
plt.figure(figsize=(8, 4))
ax = sns.barplot(x="threads_num", 
                 y="total_time", 
#                  hue="threads_num", 
                 data=dynamic8, 
                 alpha=0.7)

for counter, i in enumerate(dynamic8.index):
    if counter == 0: 
        continue 
    row = dynamic8[dynamic8.index==i]
    ax.text(counter-1.2, 
            y=row.total_time.iloc[0]+0.3, 
            s=round(row.total_time.iloc[0], 2),
            fontsize=12)
```
